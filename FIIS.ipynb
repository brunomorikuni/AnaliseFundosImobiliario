```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader.data as web
import requests
```

```python
url = r"https://www.fundsexplorer.com.br/ranking"
```

```python
web = requests.get(url)
if web.status_code == 200:
    df = pd.read_html(web.content, encoding="utf-8")[0]
```

```python
df.sort_values("Códigodo fundo", inplace = True)
```

```python
df
```

```python
df["Setor"].unique()
```

```python
df.isna().sum()
```

```python
categorical_columns = ["Códigodo fundo","Setor"]
```

```python
idx = df[df["Setor"].isna()].index
df.drop(idx, inplace = True)
```

```python
df[categorical_columns] = df[categorical_columns].astype("category")
```

```python
columns_float = list(df.iloc[:,2:-1].columns)
```

```python
df[columns_float] = df[columns_float].fillna(value = 0)
```

```python
df[columns_float]=df[columns_float].applymap(lambda x : str (x).replace("R$","").replace(".0","").replace("%","").replace('.','').replace(",",".").replace("-",""))
```

```python
df
```

```python
df[columns_float] = df[columns_float].astype('float')
```

```python
df["P/VPA"]=df["P/VPA"]/100
```

```python
df.describe()
```

```python
df.columns
```

```python
indicadores = ["Códigodo fundo",
              "Setor",
              "DY (12M)Acumulado",
              "VacânciaFísica",
              "VacânciaFinanceira",
              "P/VPA",
              "QuantidadeAtivos",
              "Liquidez Diária"]
```

```python
df_aux = df[indicadores]
```

```python
df_aux
```

```python
df["P/VPA"]
```

```python
media_setor = df_aux.groupby("Setor").agg(['mean','std'])
```

```python
media_setor
```

```python
def oportunidade_media_setor(df, setor = "Shoppings", lavel_setor = "Setor"):
    media_setor = df_aux.groupby("Setor").agg(["mean","std"])
    df_setor = df[df[lavel_setor].isin([setor])]
    
    filter_=\
            (df_setor["QuantidadeAtivos"] > 5) &\
            (df_setor["Liquidez Diária"] > 5000) &\
            (df_setor["P/VPA"] < 1.0) &\
            (df_setor["DY (12M)Acumulado"] > media_setor.loc[setor, ('DY (12M)Acumulado',"mean")])
    return df_setor[filter_]
```

```python
oportunidade = oportunidade_media_setor(df_aux, setor = "Hotel")
oportunidade.sort_values("DY (12M)Acumulado",ascending=False, inplace = True)
oportunidade
```
